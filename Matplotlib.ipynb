# Marker

| Marker | Description |
|---|---|
| `o` | Circle |
| `^` | Triangle up |
| `v` | Triangle down |
| `<` | Triangle left |
| `>` | Triangle right |
| `s` | Square |
| `p` | Pentagon |
| `*` | Star |
| `h` | Hexagon |
| `x` | Cross |
| `+` | Plus |
| `D` | Diamond |
| `d` | Thin diamond |
| `|` | Vertical bar |
| `_` | Horizontal bar |


```python
plt.plot(ypoints, marker = 'o')
```

```python
plt.plot(xpoints, marker = '*', color = 'red')
```

```python
plt.plot(xpoints, marker = '.', color = 'green')
```

```python
plt.plot(xpoints, marker = ',', color = 'blue')
```

```python
plt.plot(xpoints, marker = 'D', color = 'red')
```

```python
plt.plot(xpoints, marker = 'x', color = 'red')
```

```python
plt.plot(xpoints, marker = 'X', color = 'yellow')
```

```python
plt.plot(xpoints, marker = '+', color = 'red')
```

# Line

| Line Style | Description |
|---|---|
| `-` | Solid line |
| `--` | Dashed line |
| `-.` | Dash-dot line |
| ':' | Dotted line |
| `None` | No line |

```python
ypoints
```

```python
plt.plot(ypoints, 'o:g')

'''

 here 'o' = marker
      ':' = line reference
      'g' = color
 

'''
```

# LABEL

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
```

```python
plt.plot(x,y, '-.')

plt.title('Sports Watch Data')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
plt.plot(x,y, '-.')

plt.title('Sports Watch Data', loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```
